```python
import librosa as lb
import librosa.display
import librosa.util
import numpy as np
import matplotlib.pyplot as plt
import os
from IPython.display import Audio
```

```python
path = os.getcwd()

signal, sr = lb.load(path+'\\minidataset\\MIDI-UNPROCESSED_14-15_R1_2014_MID--AUDIO_15_R1_2014_wav--4.wav', duration = 8)
Audio(signal, rate=sr)
```

```python
mfccs = lb.feature.mfcc(signal, sr = sr, n_mfcc =127)
```

```python
inverse = librosa.feature.inverse.mfcc_to_audio(mfccs)
Audio(inverse, rate=sr)
```
